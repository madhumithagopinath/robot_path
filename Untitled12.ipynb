```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'pencil.png'  # Replace with your image file name
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold to find obstacles (white = obstacle)
_, thresh = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY)

# Robot path simulation setup
h, w = thresh.shape
robot_path = []

# Start at left-center
x, y = 0, h // 2
robot_path.append((x, y))

# Robot walking step sizes
step_x = 5
step_y = 10

# Walk across image
for x in range(0, w - step_x, step_x):
    # Check if obstacle ahead
    if thresh[y, x + step_x] == 255:
        moved = False

        # Try moving down
        if y + step_y < h and thresh[y + step_y, x + step_x] != 255:
            y += step_y
            moved = True

        # Try moving up if can't go down
        elif y - step_y > 0 and thresh[y - step_y, x + step_x] != 255:
            y -= step_y
            moved = True

        # If stuck, stop moving
        if not moved:
            print("ðŸš§ Robot stuck at:", x, y)
            break

    # Append next valid step
    robot_path.append((x + step_x, y))

# Draw robot path on image
for pt in robot_path:
    cv2.circle(img_rgb, pt, 2, (0, 255, 0), -1)  # Green dot for robot path

# Show final image
plt.figure(figsize=(12, 6))
plt.imshow(img_rgb)
plt.title("ðŸ¤– Robot Path Detection with Obstacle Avoidance")
plt.axis('off')
plt.show()
```
